# Respuestas

c)
Antes de definir la funcin principal que llevar a cabo la tarea de centralizar la matriz, definimos dos funciones secundarias para calcular la media y el desvo por columna.

Una vez hecho esto definimos la funcin principal centralizar(), que toma una matriz y obtiene los vectores con la media y el desvo llamando a las funciones anteriores. Luego resta a cada fila de la misma la media y divide por el desvo, obtieniendo as la matriz centralizada.

Adems definimos otra funcin, centralizar_test(), con el objetivo de posteriormente utilizarla para centralizar los datos de test con la media y desvo de los datos de entrenamiento.

La primera componente principal en PCA se elige para capturar la mayor cantidad de varianza en los datos. Por este motivo, centralizar es un paso muy importante, pues al restar la media a los datos se obtiene un conjunto de datos de promedio igual a 0, de manera que se elimina el sesgo en la media.
Si no se centralizan los datos, la primera componente principal se vera influenciada fuertemente por la media y no tanto por la variabilidad intrnseca de los datos.

Por otro lado, los datos deben ser divididos en cada columna por su desviacin estndar con el objetivo de estandarizar o normalizar las escalas de las variables. Si las variables tienen unidades o escalas diferentes, el anlisis de PCA podra verse afectado por aquellas variables que tienen magnitudes ms grandes. Al dividir por la desviacin estndar, todas las variables quedan en la misma escala, lo que garantiza que cada variable tenga un peso comparable en el anlisis. Esto adems facilita la interpretacin de las cargas de las variables en las componentes principales, es decir, cmo contribuyen las variables a cada componente principal.


d)

La matriz de covarianza es una matriz que contiene la covarianza entre pares de variables.

 Su dimensin depende de la cantidad de variables presentes y tendr igual cantidad de filas que de columnas, es decir, es cuadrada. En este caso, como tenemos 13 variables, la matriz de covarianza es de 13x13.
Esta matriz cuenta con las siguientes propiedades:
1. Es simtrica, la covarianza entre dos variables en la posicin i j es la misma que para esas mismas variables en la posicin j i.
1.  Cuando i=j se obtiene que la covarianza entre una variable y ella misma es la varianza. Es este caso, como adems estamos dividiendo por el desvo en cada columna, se obtiene el desvo al cuadrado en el denominador el cual se cancela dando como resultado 1 en la diagonal.
2. A partir de esta matriz se pueden calcular los componentes principales. Los autovectores de esta matriz corresponden a los autovectores de la matriz inicial.

f)

La cantidad de autovalores coincide con la dimensin de la matriz de covarianza, la cual a su vez es igual a la cantidad de variables. Cada autovalor de la matriz de covarianza representa la cantidad de varianza atribuda cada una de las componentes principales, las cuales corresponden a diferentes direcciones en el espacio de las variables.

Por otra parte, los autovectores especifican la relacin lineal entre las variables originales y las componentes principales. Los elementos del i-simo autovector corresponden a los coeficientes que indican cmo se ponderan las variables originales en la i-sima componente principal.
De esta manera, los autovectores obtenidos mediante la matriz de covarianza representan las direcciones de las componentes principales a lo largo de las cuales los datos tienen la mxima variabilidad.

```python
# A continuacin definimos las funciones que se emplearn para la realizacin de este trabajo:

# c)
# 1. media(): toma una matriz y devuelve un vector con la media de cada columna de la matriz.
def media(matriz):
  n, c= matriz.shape
  sumas= [0]*c
  for i in range(n):
    for j in range(c):
      sumas[j]+= matriz[i][j]
  media_array = [element/ n for element in sumas]
  return media_array


# 2. desvio(): toma una matriz y devuelve un vector con el desvo de cada columna de la matriz.
def desvio(matriz):
  n, c= matriz.shape
  m= media(matriz)
  sumasCuad= [0]*c
  for i in range(n):
    for j in range(c):
      sumasCuad[j]+=(matriz[i][j]-m[j])**2
  desvio_array = [np.sqrt(element/ n) for element in sumasCuad]
  #desvio_array= [np.sqrt(suma) for suma in desvio_array]
  return desvio_array

# 3. distancia(): calcula la distancia eucldea entre dos vectores.
def distancia(x1, x2):
  sumaComponentes= 0
  for i in range(len(x1)):
    sumaComponentes+= (x1[i]-x2[i])**2
  return np.sqrt(sumaComponentes)


# FUNCIONES PRINCIPALES

# centralizar(): toma una matriz, resta a cada observacin la media y divide por el desvo correspondiente
# a cada columna por separado. Se obtiene as la matriz centralizada.

def centralizar(data_array):
    n, c = data_array.shape  # n= cantidad categorias

    m= media(data_array)
    s= desvio(data_array)

    X= np.copy(data_array)

    for i in range(n):
        X[i] = X[i]- m

    for i in range(n):
        X[i] = X[i]/s

    print(type(X))

    return X, m , s

# centralizar_test(): similar a la anterior pero en lugar de calcular la media y el desvo de la matriz, se obtiene
# estos por pasaje de parmetros. Se la define de este modo con el objetivo de poder centralizar los datos de test
# correctamente con la media y desvo de los datos de entrenamiento.

def centralizar_test(data_array, m, s):
    n, c = data_array.shape  # n= cantidad categorias

    X= np.copy(data_array)

    for i in range(n):
        X[i] = X[i]- m

    for i in range(n):
        X[i] = X[i]/s

    return X

# e)

# e)
# potencia(): Retorna el atovalor de mas alto valor con su respectivo autovector
# Parmetros:
#        1. matriz: matriz centralizada.
#        2. v0: vector inicial para la iteracin.
#        3. epsilon: error minimo entre el vector el vector solucin y el aproximado.

def potencia(matriz, v0=[], epsilon=1e-8, verbose=False):
    if matriz.shape[0] != len(v0):
        print("dim(v0) != dim(A)")
        return 0
    if len(v0) == 0: v0 = 2 * np.random.rand(matriz.shape[0],1) - 1
    r_k =v0.transpose()@matriz@v0/(v0.transpose()@v0) # cociente de raleigh
    v_k = matriz@v0/np.linalg.norm(matriz@v0,2)
    delta = np.abs(v_k.transpose()@matriz@v_k/(v_k.transpose()@v_k) - r_k)

    while delta >= epsilon:
        v_k = matriz@v_k
        v_k/=np.linalg.norm(v_k,2)
        r_temp = v_k.transpose()@matriz@v_k/(v_k.transpose()@v_k)
        delta = np.abs(r_temp-r_k) # otra forma|| v_(k+1) - v_k ||_2
        r_k = r_temp
    r_k = r_k[0][0]
    #v_k = v_k.transpose()[0]
    if verbose :
        print("autovalor : ", r_k)
        print("autovector : \n", v_k)
    return r_k, v_k

# f)

def potenciaTrasladada(matriz, v0=[], cant_componentes=87 , precision=1e-6, verbose=False):
    n = matriz.dtype
    if matriz.dtype != 'float64' : n = np.float64
    matriz_c = matriz.astype(n)
    n = matriz.shape[0]
    if cant_componentes > n: cant_componentes = n
    if len(v0) == 0: v0 = 2 * np.random.rand(n,1) - 1
    d_aval =[]
    v_avec = []
    while cant_componentes > 0:
        a_val, a_vec = potencia(matriz_c, v0, epsilon=precision, verbose=False)
        #print("autovalor : ", a_val)
        matriz_c-= a_val*a_vec@a_vec.transpose()
        d_aval.append(a_val)
        v_avec.append(a_vec.transpose()[0])
        cant_componentes-=1
    v_avec = np.column_stack(v_avec)

    return d_aval, v_avec


# g)

# clas_KNN(): toma un vector de variables a partir de las cuales se quiere predecir
# una variable categrica distinta (en este caso, el tipo de vino).
# Parmetros:
#        1. x_test_i: vector de variables a predecir.
#        2. x_train_proy: matriz que contiene el conjunto de datos de entrenamiento proyectado
#        3. y_train: conjunto de clasificaciones correspondientes a los datos de entrenamiento
#        4. Ave: autovectores asociados a la matriz de covarianza de los datos de entrenamiento.
#        5. m: vector con la media de los datos de entrenamiento por columna.
#        6. s: vector con el desvo de los datos de entrenamiento por columna.
#        7. n_comp: cantidad de componentes principales que se quieren utilizar para realizar la
#        prediccin
#        8. k: cantidad de 'vecinos' ms cercanos que se quieren considerar.

def clas_KNN(x_test_i, x_train_proy, y_train, AVe, m, s, n_comp, k):

    # Centro los datos del test con la media y desvo de los datos de entrenamiento:
    x_test_cen_i = x_test_i - m
    x_test_cen_i= x_test_cen_i / s

    # Proyecto los datos segn la cantidad de componentes que tomo:
    #x_train_proy= x_train_cen @ AVe[:,0:n_comp] # CK
    x_test_proy_i= x_test_cen_i @ AVe[:,0:n_comp]

    tuplas_norma_iD = []

    # Calculo las distancias eucldeas de x_train_proy respecto a x_test_proy
    for i in range(len(x_train_proy)):
      normaI= distancia(x_train_proy[i], x_test_proy_i)
      tuplai= (i, normaI)
      tuplas_norma_iD.append(tuplai)

    # Ordeno la lista de tuplas de manera creciente segn la norma
    tuplas_ordenadas = sorted(tuplas_norma_iD, key=lambda x: x[1])

    clasificacionesDeLosK = []
    for i in range(k):
        id = tuplas_ordenadas[i][0]
        clasificacionesDeLosK.append(y_train[id])

    cantidades= {}
    cantidades[1]= clasificacionesDeLosK.count(1)
    cantidades[2]= clasificacionesDeLosK.count(2)
    cantidades[3]= clasificacionesDeLosK.count(3)

    max_key = max(cantidades, key=cantidades.get)

    return (max_key)
```
